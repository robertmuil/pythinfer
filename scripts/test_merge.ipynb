```python
from pythinfer import inout, merge
from pathlib import Path
SCRIPT_DIR = Path.cwd()
PROJ_DIR = SCRIPT_DIR.parent
SCRIPT_DIR
```

```python
cfg = inout.Project.from_yaml(PROJ_DIR / "example_projects/eg1-ancestors/ancestors.yaml")
cfg
```

```python
import pandas as pd
import rdflib

external, internal, data = merge.merge_graphs(cfg)
rows = []
for label, g in (("external", external), ("internal", internal), ("data", data)):
    if isinstance(g, rdflib.ConjunctiveGraph):
        total = 0
        for ctx in g.contexts():
            cnt = len(ctx)
            rows.append({"source": label, "graph": str(ctx.identifier), "triples": cnt})
            total += cnt
        rows.append({"source": label, "graph": "<total>", "triples": total})
    else:
        rows.append({"source": label, "graph": str(getattr(g, "identifier", label)), "triples": len(g)})

df = pd.DataFrame(rows).sort_values(["source", "graph"]).reset_index(drop=True)
display(df)
```
